## Hotel Booking Demand - (EDA)

### Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Loading Dataset

```python
df = pd.read_csv('hotel_booking_cleaned.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
# Checking missing Valus
missing_values = df.isnull().sum()
print(missing_values)

#Checking Duplicate values
duplicate_values = df.duplicated().sum()
print('Duplicate Values:-', duplicate_values)
```

### Distribution Of ADR

```python
plt.hist(df['adr'], bins=50)
plt.title('Distribution of ADR')
plt.xlabel('ADR')
plt.ylabel('Frequency')
plt.show()
```

```python
# Check Maximum ADR

print(df['adr'].max())
print(df['adr'].describe())
```

```python
#Check outliers visually

plt.boxplot(df['adr'])
plt.title('Boxplot of ADR')
plt.show()
```

```python
# Detecting and Removing outliers

#calculate Quartiles
Q1 = df['adr'].quantile(0.25)#25% of df lies below this value
Q3 = df['adr'].quantile(0.75)#75% of df lies above this value
IQR = Q3 - Q1

#calulate limits

lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR

df_no_outlier = df[
    (df['adr'] >= lower_limit) &
    (df['adr'] <= upper_limit)
]
df_no_outlier.shape

print(df_no_outlier['adr'].describe())

```

### Disribution of ADR after Outlier removal

```python
plt.hist(df_no_outlier['adr'], bins=50)
plt.title('Distribution of ADR after outlier removal')
plt.xlabel('ADR')
plt.ylabel('Frequency')
plt.show()
```

### Booking Cancellation Analysis

```python
cancellation_rate = df['is_canceled'].mean()*100
print(f"Cancellation Rate:- {cancellation_rate:.2f}%")

#Visualization

df['is_canceled'].value_counts().plot(
    kind='pie',
    autopct = '%1.1f%%'
    
)
plt.show()
```

### Hotel Type Analysis

```python
hotel_type = df['hotel'].value_counts()
hotel_type

#Visualization

df['hotel'].value_counts().plot(kind='bar')
plt.title('Booking by hotel type')
plt.show()
```

### Monthly Booking Trend

```python
monthly_booking = df['arrival_date_month'].value_counts()
monthly_booking

#Visualization

monthly_booking.plot(kind='bar')
plt.title('Monthly Booking')
plt.show()
```

### Correlation Analysis


```python
numeric_df = df.select_dtypes(include=['int64','float64'])

corr = numeric_df.corr()

import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(corr, cmap='coolwarm')
plt.show()
```

### Key Business Insights

#### 1. The hotel experiences a high cancellation rate of 27.50%, which can negatively impact revenue forecasting and room allocation.

#### 2. Booking demand is highly seasonal, with July and August recording the highest number of reservations.

#### 3. ADR analysis shows that most room prices fall within a moderate range, indicating stable pricing behavior.

#### 4. Customers with longer lead times tend to cancel more frequently, suggesting the need for targeted retention strategies.

#### 5. Demand fluctuations across months create opportunities for dynamic pricing and promotional campaigns.
