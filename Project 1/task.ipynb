## Hotel Booking Demand Task

### Importing libraries


```python
import pandas as pd
```

```python
data = pd.read_csv('hotel_bookings.csv') # Load the dataset
data.head()
data.info()
data.shape
```

```python
#checking for missing values
missing_values = data.isnull().sum()

#handling missing values
missing_persentage = (data.isnull().sum() / len(data)) * 100
missing_persentage
data.drop(['company'], axis=1, inplace=True) # Drop columns with high missing values

#Agent column
data['agent'] = data['agent'].fillna(0) # Fill missing values with 0

#Country Column 
mode_country = data['country'].mode()[0]
data['country'] = data['country'].fillna(mode_country)

#Children Column
median_children = data['children'].median()
data['children'] = data['children'].fillna(median_children)
```

```python
#Check duplicates data
duplicates = data.duplicated().sum()
print('Duplicates',duplicates)

data.drop_duplicates(inplace=True)  #drop all duplicate values
print(data.duplicated().sum())
```

```python
data.columns = data.columns.str.lower() # convert all column names into lowercase

#Convert Date Column
data['reservation_status_date'] = pd.to_datetime(  
    data['reservation_status_date']
)

#check invalid revords
print(data[data['adults'] == 0].shape)#Hotel booking without adults may be suspicious.

print(data[data['adr'] < 0])#Negative room prices are usually invalid.
```

```python
#Outlier Detection

#calculate Quartiles
Q1 = data['adr'].quantile(0.25)#25% of data lies below this value
Q3 = data['adr'].quantile(0.75)#75% of data lies below this value
IQR = Q3 - Q1

#calulate limits

lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 - 1.5*IQR

outlier = data[
    (data['adr'] < lower_limit) |
    (data['adr'] > upper_limit)
]
outlier.shape
```

```python
#Final Verification

print(data.isnull().sum())

print(data.duplicated().sum())

print(data.info())
```

```python
#Save Clean Dataset

data.to_csv(
    "hotel_booking_cleaned.csv",
    index=False
)
```
